```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

Ans-

The Bernoulli distribution is a probability distribution that represents the probability of a binary outcome, such as a coin flip or a yes/no question. 
It has only two possible outcomes, usually denoted as 0 and 1, with a probability of success (1) denoted by p and a probability of failure (0) denoted by q=1-p. 
The Bernoulli distribution is often used in modeling situations where there are only two possible outcomes, and where the outcome of one trial does not affect the outcome of subsequent trials.

An example of the Bernoulli distribution is a coin flip, where the probability of getting heads (success) is p and the probability of getting tails (failure) is q=1-p. 
Suppose we flip a fair coin, where p=0.5, the Bernoulli distribution can be used to model the probability of getting heads or tails in a single flip.

The Binomial distribution, on the other hand, is a probability distribution that represents the number of successes in a fixed number of independent trials with the same probability of success p. 
It is derived from the Bernoulli distribution by performing multiple Bernoulli trials and counting the number of successes. 
In other words, it represents the probability of getting k successes in n independent Bernoulli trials.

The key difference between the Bernoulli distribution and the Binomial distribution is that the Bernoulli distribution models the probability of a single trial with two possible outcomes, while the Binomial distribution models the probability of k successes in n independent trials with the same probability of success. 
The Bernoulli distribution is a special case of the Binomial distribution when n=1.
```

```python
Q7: Explain uniform Distribution with an example.

Ans-

In probability theory and statistics, a uniform distribution is a probability distribution where all possible outcomes are equally likely to occur.

For example, consider rolling a fair six-sided die. Each of the six possible outcomes (rolling a 1, 2, 3, 4, 5, or 6) is equally likely to occur, and thus the probability of rolling any one of them is 1/6. 
This is an example of a discrete uniform distribution, as the possible outcomes are discrete values.

Another example of a continuous uniform distribution could be the height of a person randomly selected from a group of people within a certain range. 
If the range of heights is from 5 feet to 6 feet, then any height within that range is equally likely to occur, and thus the probability density function would be a constant value of 1/(6-5)=1.

In both cases, the uniform distribution is characterized by a constant probability density function or probability mass function across the range of possible outcomes, reflecting the equal likelihood of each outcome.
```

```python
Q8: What is the z score? State the importance of the z score.

Ans-

In statistics, the z-score, also known as the standard score, is a measure of how many standard deviations an observation or data point is above or below the mean of a population. 
It is calculated by subtracting the mean of the population from the individual data point and then dividing the result by the standard deviation of the population.

The formula for calculating the z-score is:

z = (x - μ) / σ

where x is the individual data point, μ is the mean of the population, and σ is the standard deviation of the population.

The importance of the z-score is that it allows us to standardize data and compare observations from different populations or distributions on a common scale.
By converting raw data into standardized z-scores, we can easily compare observations and determine the relative position of each observation within its respective population or distribution.

The z-score is also useful in hypothesis testing and statistical inference, where it is used to determine the probability of observing a certain data point or set of data points, given the population mean and standard deviation. 
The z-score is used in many statistical analyses, such as the normal distribution, confidence intervals, and hypothesis testing.
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans-

The Central Limit Theorem is a fundamental theorem in probability theory and statistics that states that the sum or average of a large number of independent and identically distributed random variables, regardless of their distribution, tends to follow a normal distribution as the sample size increases.

In other words, if we take many random samples from any population, each with the same sample size, and compute the mean of each sample, the distribution of those sample means will be approximately normal, regardless of the underlying distribution of the population.

The significance of the Central Limit Theorem is that it allows us to make statistical inferences about a population based on a sample.
It forms the basis of many statistical techniques and methods, such as hypothesis testing, confidence intervals, and regression analysis.

Furthermore, the Central Limit Theorem plays a crucial role in quality control, as it helps us to determine whether a process is in control or out of control.
By monitoring the distribution of sample means or sample sums over time, we can detect any changes or deviations in the process and take corrective action if necessary.
```

```python
Q10: State the assumptions of the Central Limit Theorem.

Ans-

The Central Limit Theorem makes certain assumptions in order to be valid. The assumptions are as follows:

1.Independence: 
The random variables in the sample must be independent of each other.

2.Sample Size:
The sample size must be large enough to ensure that the sample mean is normally distributed.
A general rule of thumb is that the sample size should be greater than or equal to 30.

3.Identically Distributed: 
The random variables in the sample must be identically distributed, meaning that they have the same mean and variance.

4.Finite Variance:
The random variables must have a finite variance, meaning that the values cannot be too spread out.

If these assumptions are met, the Central Limit Theorem guarantees that the distribution of the sample mean will be approximately normal, regardless of the underlying distribution of the population. 
However, if the assumptions are not met, the Central Limit Theorem may not hold, and other statistical techniques may need to be used.
```
